```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_scores data.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

#droped Unnamed column.

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="Gender",data=df, color="blue")
ax.bar_label(ax.containers[0])
plt.show()
```

#from the above chart we have analysed that the number of females in the data is more than the number of males...

```python
group_by=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
group_by
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(group_by,annot=True)
plt.show()
```

#from the above chart we have concluded that the education of the parents have a good impact on their course.

```python
group_by=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
group_by
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(group_by,annot=True)
plt.show()
```

#from the above graph we have concluded that there is no negligible impact on the students's score due to their parent's marital status...

```python
plt.figure(figsize=(5,5))
sns.boxplot(x="MathScore", data=df, color="green")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x="ReadingScore", data=df, color="green")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x="WritingScore", data=df, color="green")
plt.show()
```

#from the above three charts we have concluded that the maximum students are weak in math as compared to reading and writting..
